```python
df_liyuan.dropna(subset = df_liyuan.columns[1:5], how = 'all', inplace = True)
df_liyuan.dropna(subset = ['species'], how = 'any', inplace = True)
df_liyuan.fillna(method = 'ffill', inplace = True)
df_liyuan[df_liyuan.isna().any(axis = 1)]
```

```python
df_liyuan.isna().any()
```

```python
df_liyuan[df_liyuan['sepal_length']>10]
df_liyuan.drop(df_liyuan[df_liyuan['sepal_length']>10].index, inplace = True)
df_liyuan[df_liyuan['sepal_length']>10]
```

```python
df_liyuan[df_liyuan.duplicated()]
```

```python
df_liyuan.drop_duplicates(inplace=True)
df_liyuan[df_liyuan.duplicated()]
```

```python
df_liyuan.drop(columns=['nr'], inplace=True, axis=1)
df_liyuan.head()
```

```python
import numpy as np
df_liyuan['sepal_area'] = df_liyuan.apply(
    lambda x:np.pi*(x['sepal_length']/2)*(x['sepal_width']/2),
    axis = 1
    )
df_liyuan
```

```python
species_mapping = {'setosa':0, 'versicolor':1, 'virginica':2}
df_liyuan['species'] = df_liyuan['species'].map(species_mapping)
df_liyuan.head()
```
